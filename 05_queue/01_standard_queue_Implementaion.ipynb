# 1. queue implementation using lists

```python
class QueueUsingList:
    def __init__(self) -> None:
        self.items = []
        
    def is_empty(self):
        return len(self.items) == 0
    
    def display(self):
        if self.is_empty():
            return "Queue is empty"
        print(self.items)
        
    def enqueue(self, value):
        self.items.append(value)
        self.display()
        
    def dequeue(self):
        if self.is_empty():
            return "Queue is empty"
        
        deleted = self.items.pop(0)
        self.display()
        return deleted
    
    def front_peek(self):
        if self.is_empty():
            return "Queue is empty"
        
        return self.items[0]
    
    def rear_peek(self):
        if self.is_empty():
            return "Queue is empty"
        return self.items[-1]
    
```

# 2. Queue Implementation using Linked List

```python
class Node:
    def __init__(self, data) -> None:
        self.data = data
        self.next = None
```

```python
class QueueUsingLinkedList:
    def __init__(self):
        self.rear = self.front = None
        
    def is_empty(self):
        return self.front is None
    
    def display(self):
        if self.front is None:
            return "empty queue"
        
        curr = self.front
        while curr:
            print(curr.data, end=" ")
            curr = curr.next
        print()
        
    def enqueue(self, value):
        temp = Node(value)
        
        if self.rear is None:
            self.front = self.rear = temp
        else:
            self.rear.next = temp
            self.rear = temp
        self.display()    
        
    def dequeue(self):
        if self.is_empty():
            return "Queue is Empty"
        
        if self.front == self.rear: # checking if single value is present
            self.front = self.rear = None
        else:
            self.front = self.front.next
            
        self.display()
        
        # curr = self.front
        # self.front = curr.next
        # if not self.front:
        #     self.rear = None
        # dequeued = curr.data
        # self.display()
        # return dequeued  we can use this also
        
    def front_peek(self):
        if self.is_empty():
            return "Queue is empty"
        
        return self.front.data
    
    def rear_peek(self):
        if self.is_empty():
            return "Queue is empty"
        return self.rear.data
    
    
        
```
