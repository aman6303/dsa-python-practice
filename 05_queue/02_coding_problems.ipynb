# 1. Queue using Stack

```python
from collections import deque
```

```python
class QueueUsingStack:
    def __init__(self) -> None:
        self.s1 = deque()  # assume deque as stack
        self.s2 = deque()
        
    def enqueue(self, value):
        self.s1.append(value)
        
    def dequeue(self):
        # first copy all the values from stack s1 to s2
        if not self.s2:
            while self.s1:
                self.s2.append(self.s1.pop())  # we are storing in reverse
                
        if not self.s2:
            return "empty queue"
        return self.s2.pop()
        
```

```python
my_queue = QueueUsingStack()
my_queue
```

```python
my_queue.dequeue()
```
