### 1. Count Vowels and Consonents

- assume all characters are lowercase
- string may include special characters

```python
from curses.ascii import isalpha


mystring = "my demo && strign"

v_count = 0
c_count = 0

for char in mystring:
    if char.isalpha():
        if char in ['a', 'i', 'e', 'o', 'u']:
            v_count += 1
        else:
            c_count += 1
            
print(f"{v_count} vowels and {c_count} consonants")
```

### 2. Count Frequency of Characters

```python
mystring = "this is my stringd 888"

count = dict()
for char in mystring:
    if char not in count.keys():
        count[char] = mystring.count(char)
        
print(count)

# we can do the same without using count

countdict = {}
for char in mystring:
    countdict[char] = countdict.get(char, 0) + 1
    
print(countdict)
```

### 4. Check for Anagrams

- Return `True` if two strings contain the same characters in any order
- Strings will be of same length and lowercase

```python
s1 = 'listen'
s2 = 'silent'
```

#### Time: `nlog(n)`

```python
s1_sorted = sorted(s1)
s2_sorted = sorted(s2)

if s1_sorted == s2_sorted:
    print('Anagram')
else:
    print('Not Anagram')
```

#### Time: `n`

```python
ord('a'), ord('a') - 97 
```

```python
ord('z'), ord('z') - 97
```

```python
ord('s'), ord('s') - 97
```

```python
mapping = [0] * 26

for ch in s1:
    index = ord(ch) - 97
    mapping[index] += 1

for ch in s2:
    index = ord(ch) - 97
    mapping[index] -= 1

if all(m == 0 for m in mapping):
    print('Anagram')
else:
    print('Not Anagram')
```
