### 1. Strings in Python

#### 1.1 Ways of creating a String

```python
str1 = 'This is a string in Python'
print(str1)
print(type(str1))
```

```python
str2 = "This is also a string in Python"
print(str2)
print(type(str2))
```

```python
str3 = '''This is a string within triple quotes'''
print(str3)
print(type(str3))
```

```python
str4 = """This is also a string within triple quotes"""
print(str4)
print(type(str4))
```

```python
print('Hi "John"')
```

```python
print('''This is a 
multi-line string that spans
multiple lines''')
```

```python
print('This is a \
multi-line string that spans \
multiple lines')
```

### 3. Common String Functions

```python
str1 = 'Data SCIEnce'

print(str1)
print(id(str1)) # getting id
```

#### 3.1 `lower`

```python
lower_str1 = str1.lower()

print(lower_str1)
print(id(lower_str1)) # always a new string will be created with same name (as strings are immutable)
```

```python
str1
```

#### 3.2 `upper`

```python
str1.upper()
```

```python
str2 = 'This is 123 ABC'

print(str2.lower())
print(str2.upper())
```

#### 3.3 `strip`

```python
str3 = '   data science  '
```

```python
str3.strip()
```

```python
str4 = '$$$$###data science#############@@@@@'

str4.strip('$').strip('@').strip('#')
```

#### 3.4 `replace`

```python
str1 = 'Data Science'
str1
```

```python
str1.replace('Science', 'Analysis')
```

```python
str1.replace('a', 'z')
```

```python
str2 = 'Data Science and Data Engineering'
str2
```

```python
str2.replace('Data', 'data', 1) # replace first occurance only
```

```python
str2.replace('Data', 'data', 2) # replace first two occurance only
```

#### 3.5 `split`

```python
str2
```

```python
str2.split()
```

```python
type(str2), type(str2.split())
```

```python
str3 = 'Data-Science-and-Data-Engineering'
str3
```

```python
str3.split('-')
```

```python
str3.split('-', 3)
```

```python
str2.split('Data')
```

```python
name = 'Mr. James Anderson'

# doing first split only
sal, full_name = name.split(maxsplit=1)
print(sal)
print(full_name)
```
