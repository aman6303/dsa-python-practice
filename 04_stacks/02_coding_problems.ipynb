# 1. reverse a string using stack

```python
from collections import deque
```

```python
def reverse_string_using_stack(string):
    
    stack = deque()
    
    for char in string:
        stack.append(char)
    
    ans = ""
    # for i in range(len(stack)):
    #     ans += stack.pop()
    while stack:
        ans += stack.pop()
    return ans
```

```python
reverse_string_using_stack('This is a string!!')
```

# 2. check balanced paranthesis

```python
def is_balanced_parantheses(expression):
    stack = deque()
    
    pairs = {
        ')' : '(',
        '}' : '{',
        ']' : '['
    }
    
    for exp in expression:
        if exp in pairs.values():
            stack.append(exp)
            
        elif exp in pairs.keys():
            pair = pairs[exp]
            
            if (not stack) or pair != stack.pop():
                return False
            
    return len(stack) == 0
            
```

```python
is_balanced_parantheses('(({[{[]()[]{}}]}))')
```

# 3. Undo operation simulation

```python
def simulate_undo_operation(commands):
    
    stack = deque()
    
    for command in commands:
        if command == 'undo':
            if stack:
                stack.pop()
            else:
                print("stack is empty")
        else:
            stack.append(command[-1])
            
    return "".join(stack)
```

```python
commands = [
    'type A',
    'type B',
    'undo',
    'type C',
    'type C',
    'type F'
]

res = simulate_undo_operation(commands)
print(f'Result: {res}')
```
