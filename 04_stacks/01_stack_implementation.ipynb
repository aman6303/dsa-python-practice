# 1. stack Implementation using list

- get_size
- is_empty
- display
- push
- pop
- peek

```python
class MyStack:
    def __init__(self) -> None:
        self.stack = []

    def get_size(self):
        return len(self.stack)
    
    def is_empty(self):
        if self.get_size() == 0:
            return True
        return False
    
    def display(self):
        if self.is_empty():
            return "empty list"
        for i in self.stack[::-1]:
            print(i)
        print()
        
    def push(self, value):
        self.stack.append(value)
        self.display()
        
    def pop(self):
        if self.is_empty():
            return "empty list"
        
        return self.stack.pop()
    
    def peek(self):
        if self.is_empty():
            return "empty list"
        
        return self.stack[-1]
    
    
```

# 2. Stack implementation using linked list

- is_empty
- display
- push
- pop
- peek

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None
        
```

```python
class StackUsingLinkedList:
    def __init__(self):
        self.head = None
        self.length = 0
        
    def is_empty(self):
        if self.length == 0:
            return True
        return False
    
    def display(self):
        if self.is_empty():
            return "empty stack"
        
        curr = self.head
        while curr:
            print(curr.data)
            curr = curr.next
        print()
        
    def push(self, value):
        temp = Node(value)
        
        if self.is_empty():
            self.head = temp
        else:
            temp.next = self.head
            self.head = temp
            
        self.length += 1
        self.display()
        
    def pop(self):
        if self.is_empty():
            return "empty list"
        
        if self.length == 1:
            pop_element = self.head.data
            self.head = None
        else:
            pop_element = self.head.data
            self.head = self.head.next
        self.length += 1
        self.display()
        return pop_element
        
    def peek(self):
        if self.is_empty():
            return "empty stack"
        return self.head.data
    

        
```

```python
my_stack = StackUsingLinkedList()
```
