# 1. Circular Singly Linked list

- is_empty
- display
- insert
- search 
- delete

```python

class Node:
    def __init__(self, data) -> None:
        self.data = data
        self.next = None
```

```python
class CircularSLL:
    def __init__(self) -> None:
        self.head = None
        self.length = 0
        
    def is_empty(self):
        if self.head is None:
            return True
        return False
    
    def display(self):
        if self.is_empty():
            print('Empty list')
        else:
            current = self.head
            while True:
                print(current.data, end=' ')
                current = current.next
                if current == self.head:
                    break
            print()
            
    def insert_at_beginning(self, value):
        temp = Node(value)
        if self.is_empty():
            self.head = temp
            temp.next = self.head
        else:
            curr = self.head
            
            while curr.next != self.head:
                curr = curr.next
                
            temp.next = self.head
            self.head = temp
            curr.next = temp
        self.length += 1
        self.display()
        
    def insert_at_end(self, value):
        temp = Node(value)
        
        if self.is_empty():
            self.head = temp
            temp.next = self.head
            
        else:
            curr = self.head
            
            while curr.next != self.head:
                curr = curr.next
                
            curr.next = temp
            temp.next = self.head
        
        self.length += 1    
        self.display()
        
        
    def insert_at_middle(self, pos, value):
        if pos == 0:
            self.insert_at_beginning(value)
        elif pos == self.length:
            self.insert_at_end(value)
        elif 0 < pos < self.length-1:
            temp = Node(value)
            
            curr = self.head
            for _ in range(pos-1):
                curr = curr.next
                
            temp.next = curr.next
            curr.next = temp
            self.length += 1
            self.display()
        else:
            return "Invalid position"
        
    def search(self, value):
        curr = self.head
        
        if curr is None:
            return "empty list"
        pos = 0
        while True:
            if curr.data == value:
                return f"found at {pos}"
            pos += 1
            curr = curr.next
            if curr == self.head:
                break
        return "not found"
    
    def delete_at_beginning(self):
        if self.is_empty():
            print('Cannot delete from empty list')
            return

        if self.length == 1:
            self.head = None
        else:
            current = self.head
            while current.next != self.head:
                current = current.next
            self.head = self.head.next
            current.next = self.head
        self.length -= 1
        self.display()
            
        
    def delete_at_end(self):
        if self.is_empty():
            return "list is empty"
        
        if self.length == 1:
            self.head = None
        else:
            p = self.head
            q = None
            while p.next != self.head:
                q = p
                p = p.next
                
            q.next = self.head
            p.next = None
            
        self.length -= 1
        self.display()
        
    def del_at_middle(self, pos):
        if pos == 0:
            self.delete_at_beginning()
        elif pos == self.length - 1:
            self.delete_at_end()
        elif 0 < pos < self.length-1:
            p = self.head
            q = None
            for _ in range(pos):
                q = p
                p = p.next  
            q.next = p.next
            p.next = None
            self.length -= 1
            self.display()
        else:
            return "Invalid pos"
            
        
        
```
