# 1. Circular Singly Linked list

- is_empty
- display
- insert
- search 
- delete

```python

class Node:
    def __init__(self, data) -> None:
        self.data = data
        self.next = None
```

```python
class CircularSLL:
    def __init__(self) -> None:
        self.head = None
        self.length = 0
        
    def is_empty(self):
        if self.head is None:
            return True
        return False
    
    def display(self):
        if self.is_empty():
            print('Empty list')
        else:
            current = self.head
            while True:
                print(current.data, end=' ')
                current = current.next
                if current == self.head:
                    break
            print()
            
    def insert_at_beginning(self, value):
        temp = Node(value)
        if self.is_empty():
            self.head = temp
            temp.next = self.head
        else:
            curr = self.head
            
            while curr.next != self.head:
                curr = curr.next
                
            temp.next = self.head
            self.head = temp
            curr.next = temp
        self.length += 1
        self.display()
        
    def insert_at_end(self, value):
        temp = Node(value)
        
        if self.is_empty():
            self.head = temp
            temp.next = self.head
            
        else:
            curr = self.head
            
            while curr.next != self.head:
                curr = curr.next
                
            curr.next = temp
            temp.next = self.head
        
        self.length += 1    
        self.display()
        
        
    def insert_at_middle(self, pos, value):
        if pos == 0:
            self.insert_at_beginning(value)
        elif pos == self.length:
            self.insert_at_end(value)
        elif 0 < pos < self.length-1:
            temp = Node(value)
            
            curr = self.head
            for _ in range(pos-1):
                curr = curr.next
                
            temp.next = curr.next
            curr.next = temp
            self.length += 1
            self.display()
        else:
            return "Invalid position"
        
    def search(self, value):
        curr = self.head
        
        if curr is None:
            return "empty list"
        pos = 0
        while True:
            if curr.data == value:
                return f"found at {pos}"
            pos += 1
            curr = curr.next
            if curr == self.head:
                break
        return "not found"
    
    def delete_at_beginning(self):
        if self.is_empty():
            print('Cannot delete from empty list')
            return

        if self.length == 1:
            self.head = None
        else:
            current = self.head
            while current.next != self.head:
                current = current.next
            self.head = self.head.next
            current.next = self.head
        self.length -= 1
        self.display()
            
        
    def delete_at_end(self):
        if self.is_empty():
            return "list is empty"
        
        if self.length == 1:
            self.head = None
        else:
            p = self.head
            q = None
            while p.next != self.head:
                q = p
                p = p.next
                
            q.next = self.head
            p.next = None
            
        self.length -= 1
        self.display()
        
    def del_at_middle(self, pos):
        if pos == 0:
            self.delete_at_beginning()
        elif pos == self.length - 1:
            self.delete_at_end()
        elif 0 < pos < self.length-1:
            p = self.head
            q = None
            for _ in range(pos):
                q = p
                p = p.next  
            q.next = p.next
            p.next = None
            self.length -= 1
            self.display()
        else:
            return "Invalid pos"
            
        
        
```

# 2. Circular Doubly linked list

- is_empty
- display (forward and backword)
- insert (end, begin and middle)
- search
- delete(end, begin and middle) 

```python
class Node:
    def __init__(self, data) -> None:
        self.data = data
        self.next = None
        self.prev = None
```

```python
class MyCircularDoublyLinkedList:
    def __init__(self) -> None:
        self.head = None
        self.tail = None
        self.length = 0
        
    def is_empty(self):
        if self.head is None:
            return True
        return False
    
    def display_forward(self):
        if self.is_empty():
            return "empty list"
        
        curr = self.head
        
        while True:
            print(curr.data, end = " ")
            curr = curr.next
            
            if curr == self.head:
                break
            
    def display_backward(self):
        if self.is_empty():
            return "empty list"
        
        curr = self.tail
        
        while True:
            print(curr.data, end = " ")
            curr = curr.prev
            
            if curr == self.tail:
                break
        
    def insert_at_beginning(self, value):
        temp = Node(value)
        
        if self.is_empty():
            self.head = temp
            self.tail = temp
            temp.next = temp  # putting the same address is each node
            temp.prev = temp
            
        else:
            temp.next = self.head
            temp.prev = self.tail
            self.head.prev = temp
            self.head = temp
            self.tail.next = temp
            
        self.length += 1
        self.display_forward()
        
    def insert_at_end(self, value):
        temp = Node(value)
        
        if self.is_empty():
            self.head = temp
            self.tail = temp
            temp.next = temp 
            temp.prev = temp
        else:
            temp.prev = self.tail
            temp.next = self.head
            self.tail.next = temp
            self.tail = temp
            self.head.prev = temp
            
        self.length += 1
        self.display_forward()
        
    def insert_at_middle(self, pos, value):
        if pos == 0:
            self.insert_at_beginning(value)
        elif pos == self.length:
            self.insert_at_end(value)
        elif 0 < pos < self.length:
            curr = self.head
            for _ in range(pos):
                curr = curr.next
            
            temp = Node(value)
            temp.next = curr
            temp.prev = curr.prev
            curr.prev.next = temp
            curr.prev = temp
            self.length += 1
            self.display_forward()
        else:
            return "Invalid Pos"
        
    def search(self, value):
        if self.is_empty():
            return "empty list"
        curr = self.head
        pos = 0
        while True:
            if curr.data == value:
                return f"found at {pos}"
            curr = curr.next
            pos += 1
            
            if curr == self.head:
                break
            
        return "not found"
    
    def delete_at_beginning(self):
        if self.is_empty():
            return "empty list"
        
        if self.length == 1:
            self.head = None
            self.tail = None
        else:
            self.head = self.head.next
            self.head.prev = self.tail
            self.tail.next = self.head
        self.length -= 1
        self.display_forward()
        
    def delete_at_end(self):
        if self.is_empty():
            return "empty list"
        
        if self.length == 1:
            self.head = None
            self.tail = None
        else:
            self.tail = self.tail.prev
            self.tail.next = self.head
            self.head.prev = self.tail
            
        self.length -= 1
        self.display_forward()
        
    def delete_at_middle(self, pos):
        if pos == 0:
            self.delete_at_beginning()
        elif pos == self.length-1:
            self.delete_at_end()
        elif 0 < pos < self.length-1:
            curr = self.head
            for _ in range(pos):
                curr = curr.next
            
            q = curr.prev
            q.next = curr.next
            curr.next.prev = q
            curr.next = curr.prev = None
            self.length -=1
            self.display_forward()
        else:
            return "Invalid Index"
    
```
