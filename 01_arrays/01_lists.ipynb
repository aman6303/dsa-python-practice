### 1.3 Insertion in lists

#### 1.3.1 using append

```python
arr1 = [2,3,4,5,5,4, "ama"]
arr1.append("aman")
arr1
```

#### 1.3.2 using insert

```python
arr1.insert(0, "element")
arr1
```

### 1.4 merging two lists

#### 1.4.1 using extend

```python
arr1 = ["1", '20']
arr2 = ['ele1', 'ele2']
arr1.extend(arr2)
arr2.extend(arr1)  # here we are updating arr2 with updated arr1

arr1, arr2
```

#### 1.4.2 using concatination

```python
arr1 + arr2
```
